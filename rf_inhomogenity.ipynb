```python
from __future__ import division
%pylab inline
from qinfer.distributions import ProductDistribution, NormalDistribution, InterpolatedUnivariateDistribution
from scipy.linalg import expm
```

```python
def asym_lorentz_scale(x, x_0, gamma_0, a):
    return 2 * gamma_0 / (1 + np.exp(a * (x - x_0)))
```

```python
def asym_lorentz_pdf(x, x_0, gamma_0, a):
    gamma = asym_lorentz_scale(x, x_0, gamma_0, a)
    return 2 * gamma / (np.pi * gamma_0 * (1 + 4 * ((x - x_0) / (gamma_0))**2))
```

```python
dist = ProductDistribution(
    NormalDistribution(1, 0.075**2),
    InterpolatedUnivariateDistribution(lambda x: asym_lorentz_pdf(x, 10, 2, 2), 2, 1200)
)
```

```python
samples = dist.sample(n=10000)
figure(figsize=(12, 6))
subplot(1, 2, 1)
hist(samples[:, 0], bins=100);
subplot(1, 2, 2)
hist(samples[:, 1], bins=100);
```

```python
X = np.array([[0, 1], [1, 0]])
Z = np.array([[1, 0], [0, -1]])
```

```python
@np.vectorize
def cond_signal(t, tau, t_p, delta_w, w_rf):
    H = delta_w * Z
    H_p = delta_w * Z + w_rf * X
    psi = np.array([[1], [1]]) / np.sqrt(2)
    U1 = expm(-1j * tau * H)
    U2 = expm(-1j * t_p * H_p)
    U3 = expm(-1j * t * H)
        
    return np.abs(reduce(np.dot, [
        psi.T, U3, U2, U1, psi
    ]))**2
```

```python
ts = linspace(0, 20, 100)
t_p = 0.155
sig = cond_signal(ts, 4, t_p, 1, 10)
plot(ts, sig)
ylim(0, 1)
```
